```python
import pandas as pd
```

```python
df = pd.DataFrame({'city':['New York','London','Paris','Tokyo','New York','Paris'],
                   'price' : [200,150,300,250,180,320]})
```

```python
df.head()
```

```python
##calculate mean for each city
mean_price = df.groupby('city').mean().to_dict()
```

```python
##replace each city with mean:
mean_price
```

```python
df['city_enoded'] = df['city'].map(mean_price['price'])
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
mean_price = df.groupby('total_bill').mean(numeric_only=True).to_dict() #The numeric_only parameter tells pandas to only calculate the mean of numeric columns.
```

```python
mean_price
```

```python
df['sex_encoded'] = df['sex'].map(mean_price['tip'])
```
