```python
from sklearn.datasets import make_classification
```

```python
##X = independent feature
##y = dependend output feature
X,y = make_classification(n_samples = 1000,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
import pandas as pd

df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 =pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
##tranform the data set:
oversample = SMOTE()
X_over,y_over = oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
#
```

```python

df1 = pd.DataFrame(X_over,columns=['f1','f2'])
df2 =pd.DataFrame(y_over,columns=['target'])
oversample_df = pd.concat([df1,df2],axis=1)
```
