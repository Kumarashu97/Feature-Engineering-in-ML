```python
## 1.linear interpolation:

import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
##interpolation the data using linear intrpolation
x_new = np.linspace(1,5,100)
y_new = np.interp(x_new,x,y)
```

```python
print(y_new)
```

```python
plt.scatter(x_new,y_new)
plt.show()
```

```python
## 2 cubic interpolation using scipy
from scipy.interpolate import interp1d
cs = interp1d(x,y,kind='cubic')
y_new = cs(x_new)
plt.scatter(x_new,y_new)
plt.show()
```

```python
## 2 quadratic interpolation using scipy
from scipy.interpolate import interp1d
cs = interp1d(x,y,kind='quadratic')
y_new = cs(x_new)
plt.scatter(x_new,y_new)
plt.show()
```

```python
p = np.polyfit(x,y,2)
print(p)
```

```python
x_new = np.linspace(1,5,100)
y_new = np.polyval(p,x_new)
plt.scatter(x_new,y_new)
plt.show()
```
